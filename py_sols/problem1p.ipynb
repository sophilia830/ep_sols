```python
#Find the sum of all the multiples of 3 or 5 below 1000.

import time

fact1 = int(input("Give the first factor."))
fact2 = int(input("Give the second factor."))
fact3 = fact1*fact2
below = int(input("Factors 1 and 2 are multiples of: "))

#METHOD 1

tstart = time.perf_counter()

init = 0
sum = 0

while init < below:
    if (init%fact1 == 0) or (init%fact2 == 0):
        sum += init
    init += 1

tend = time.perf_counter()
runtime = str(tend - tstart)

output = f"\nThe sum of the factors of either {fact1} or {fact2} or both that are less than {below} is {sum}.\n"
print(output)
print("Method 1: " + runtime + " seconds")

#METHOD 2

tstart = time.perf_counter()

div1 = int(below/fact1)
if(below%fact1 == 0):
    div1 += -1

div2 = int(below/fact2)
if(below%fact2 == 0):
    div2 += -1

div3 = int(below/(fact1*fact2))
if(below%fact3 == 0):
    div3 += -1

sum1 = 0
sum2 = 0
sum3 = 0

for i in range(div1 + 1):
    sum1 += fact1*i

for i in range(div2 + 1):
    sum2 += fact2*i

for i in range(div3 + 1):
    sum3 += fact3*i

sum = sum1 + sum2 - sum3

tend = time.perf_counter()
runtime = str(tend - tstart)

#print(sum)
print("Method 2: " + runtime + " seconds")

    


    
```
